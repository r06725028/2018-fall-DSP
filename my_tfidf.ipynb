```python
import pandas as pd
import numpy as np
import nltk,re
from nltk import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from pprint import pprint
from collections import Counter
from math import *
import matplotlib.pyplot as plt 
```

```python
#set variables
ROOT="IRTM/"
N=1000
```

```python
#get with iterate
def get_text_data(index):
    with open(ROOT+str(index)+".txt","r") as f:
        text_data=f.read()
    pprint(str(index)+" : "+text_data[:10])
    return(text_data)
```

```python
#using stop words in nltk corpus
stop_words=(set(stopwords.words("english")))


def tokenize_to_list(text):
    out_list=list()
    #using nltk tokenizer
    tokenized_text=(word_tokenize(text))
    #using nltk PorterStemmer
    ps = PorterStemmer()
    #iteration
    for word in tokenized_text:
        #stemming and lower case
        word=ps.stem(word.lower())
        # word = ps.stem(word) 
        if word not in stop_words:
            out_list.append(word)
    return(out_list)
#define function get_tf_df() as follow
#use collection Counter to get tf for each doc
#use collection Counter to update df
def get_tf_df(text_data,df=None):
    text_data = re.sub('[#?|$|.|!|\'|0-9|\n+|,|"|`|\-+|/|_|*]|@','',text_data)
    text_data_list=tokenize_to_list(text_data)
    tf=Counter()
    if not df:
        df=Counter()
    tf.update(text_data_list)
    df.update(set(text_data_list))
    return(tf,df)
```

```python
#do iteration
tf_list=list()
df=Counter()
for i in range(1,N+1):
    text_data=get_text_data(i)
    tf,df=update_tf_df(text_data=text_data,df=df)
    tf_list.append(tf)
```

```python
#construct idf dictoinary
idf=dict({key:log10(float(N)/float(df.get(key))) for key in df.keys()})
```

```python
#construct tfidf list with the idf dict above
tfidf_list=list()
for doc_tf_dict in tf_list:
    tfidf=dict()
    for key in df.keys():
        if doc_tf_dict.__contains__(key):
            tfidf.update({key:doc_tf_dict[key]*idf[key]})
        else:
            tfidf.update({key:0})
    tfidf_list.append(tfidf)
            
```

```python
#define function cosine to return cos value of two doc index
def cosine(docx,docy):
    vectorx=np.array(list(tfidf_list[docx-1].values()))
    vectory=np.array(list(tfidf_list[docy-1].values()))
    normx=np.sqrt(vectorx.dot(vectorx))
    normy=np.sqrt(vectory.dot(vectory))
    return(np.dot(vectorx,vectory)/(normx*normy))
```

```python
#manipulate df dictionary with pandas to fit homework request
df_pd=pd.DataFrame.from_dict(df,orient='index',columns=["df"])
df_pd['term']=df_pd.index.values
df_pd.sort_values('term',inplace=True)
df_pd.set_index(pd.RangeIndex(len(df_pd)),inplace=True)
df_pd.to_csv("df_dict.csv",index_label="term id")
```

```python
#manipulate tf dictionary of doc one with pandas to fit homework request
docone_term_list=list(tf_list[0].keys())
docone_dict=df_pd.loc[df_pd.term.isin(docone_term_list),['term']]
docone_dict['tfidf']=[tfidf_list[0].get(key) for key in docone_dict.term]
docone_dict.to_csv("doc_one_tfidf.csv",index_label="term id")
```
